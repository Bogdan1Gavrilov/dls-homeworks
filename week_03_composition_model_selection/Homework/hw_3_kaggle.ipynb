# Метрика

Перед решением любой задачи важно понимать, как будет оцениваться ваше решение. В данном случае мы используем стандартную для задачи классификации метрику ROC-AUC. Ее можно вычислить, используя только предсказанные вероятности и истинные классы без конкретного порога классификации + она раотает даже если классы в данных сильно несбалансированны (примеров одного класса в десятки раз больше примеров длугого). Именно поэтому она очень удобна для соревнований.

Посчитать ее легко:


```python
from sklearn.metrics import roc_auc_score

y_true = [
    0,
    1,
    1,
    0,
    1
]

y_predictions = [
    0.1,
    0.9,
    0.4,
    0.6,
    0.61
]

roc_auc_score(y_true, y_predictions)
```

```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
```

```python
X = data.drop('Churn', axis=1)
y = data['Churn']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42, stratify=y)#Используем stratify чтобы
#распределение классов в трейне и тесте было +- равным
```

```python
OHE = OneHotEncoder(drop='first', sparse_output=False)#Указываю sparse=false чтобы возвращались не индексы единиц а матрица целиком
X_train_cat = OHE.fit_transform(X_train[cat_cols])
X_test_cat = OHE.transform(X_test[cat_cols])

scaler = StandardScaler()
X_train_num = scaler.fit_transform(X_train[num_cols])
X_test_num = scaler.transform(X_test[num_cols])

#И объединяем с помощью hstack
X_train_preprocess = np.hstack([X_train_cat, X_train_num])#Не заменяю исходники чтобы потом для катбуста их использовать
X_test_preprocess = np.hstack([X_test_cat, X_test_num])
```

```python
model = LogisticRegressionCV(
    Cs=[0.001, 0.1, 1, 10, 100],
    cv=5,
    max_iter=1000,
    scoring='roc_auc'
)
```

```python
model.fit(X_train_preprocess, y_train)
print('Точность: ', model.score(X_test_preprocess, y_test))
print('Лучшее C: ', model.C_[0])
```
