```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
```

   robust scaler.        standart 

```python
X = data.drop('Churn', axis=1)
y = data['Churn']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42, stratify=y)# stratify 
#       +- 
```

```python
OHE = OneHotEncoder(drop='first', sparse_output=False)# sparse=false        
X_train_cat = OHE.fit_transform(X_train[cat_cols])
X_test_cat = OHE.transform(X_test[cat_cols])

scaler = StandardScaler()
X_train_num = scaler.fit_transform(X_train[num_cols])
X_test_num = scaler.transform(X_test[num_cols])

#    hstack
X_train_preprocess = np.hstack([X_train_cat, X_train_num])#        
X_test_preprocess = np.hstack([X_test_cat, X_test_num])
```

```python
model = LogisticRegressionCV(
    Cs=[0.001, 0.1, 1, 10, 100],
    cv=5,
    max_iter=1000,
    scoring='roc_auc'
)
```

```python
model.fit(X_train_preprocess, y_train)
print(': ', model.score(X_test_preprocess, y_test))
print(' C: ', model.C_[0])
```

```python
model_boost = CatBoostClassifier(eval_metric='AUC')#   ,  .       
model_boost.fit(X_train, y_train, cat_features=cat_cols)
```

```python
y_prob = model_boost.predict_proba(X_test)
```
